```python
#function to calculate inflation
inflate_dict = {'2006': 1.25, '2007': 1.21, '2008': 1.17, '2009': 1.17, '2010': 1.15, '2011': 1.12, '2012': 1.09, 
                '2013': 1.08, '2014': 1.06, '2015': 1.06, '2016': 1.05, '2017': 1.02}

def inflation(year):
    [frame.Dollars * current_inflation for current_year, current_inflation in inflate_dict.items() if year == current_year]
```

```python
#import salary data
import pandas as pd
import html5lib

frame1 = pd.read_html(r'http://www.espn.com/mlb/freeagents')
frame1 = pd.DataFrame(frame1[0])
frame1.columns = frame1.iloc[1]

#create dataframe that doesnt have PLAYER, FA, Minor Lg, and --
frame1 = frame1[frame1.PLAYER != 'PLAYER']
frame1 = frame1[frame1.STATUS != 'FA']
frame1 = frame1[frame1.DOLLARS != 'Minor Lg']
frame1 = frame1[frame1.DOLLARS != '--']

#removing upper case to install capitalization for efficient behavior
frame1.columns = [str(frame1.columns[i]).capitalize() for i in range(len(frame1.columns))]

#removing $ and , from Dollars columns, and setting datatype to int
frame1['Dollars'] = frame1['Dollars'].replace({'\$':'', ',':''}, regex=True)
frame1 = frame1[pd.to_numeric(frame1['Dollars'], errors='coerce').notnull()]
frame1['Dollars'] = frame1['Dollars'].astype(float)
    
#treat Yrs as int
frame1 = frame1[pd.to_numeric(frame1['Yrs'], errors='coerce').notnull()]
frame1['Yrs'] = frame1['Yrs'].astype(int)

#removing unneccesary row
frame1 = frame1.rename(columns={'2018 team': 'Former Team'})

#include what year
frame1['Year'] = '2018'
```

```python
#getting all free agents from 2006-2017
years = ['2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017']
dataframe_names = []

for year in range(len(years)):
    url = 'http://www.espn.com/mlb/freeagents/_/year/' + years[year]
    frame = pd.read_html(url)
    frame = pd.DataFrame(frame[0])
    frame.columns = frame.iloc[1]

    #create dataframe that doesnt have PLAYER, FA, Minor Lg, and --
    frame = frame[frame.PLAYER != 'PLAYER']
    frame = frame[frame.STATUS != 'FA']
    frame = frame[frame.DOLLARS != 'Minor Lg']
    frame = frame[frame.DOLLARS != '--']

    #removing upper case to install capitalization for efficient behavior
    frame.columns = [str(frame.columns[i]).capitalize() for i in range(len(frame.columns))]

    #removing $ and , from Dollars columns, and setting datatype to int
    frame['Dollars'] = frame['Dollars'].replace({'\$':'', ',':''}, regex=True)
    frame = frame[pd.to_numeric(frame['Dollars'], errors='coerce').notnull()]
    frame['Dollars'] = frame['Dollars'].astype(float)
    
    #treat Yrs as int
    frame = frame[pd.to_numeric(frame['Yrs'], errors='coerce').notnull()]
    frame['Yrs'] = frame['Yrs'].astype(int)
    
    #removing unneccesary row
    frame = frame.rename(columns={years[year]+' team': 'Former Team'})
    
    #include what year
    frame['Year'] = years[year]
    
    #call the inflation function
    inflation(years[year])
    
    #append dataframe_names
    dataframe_names.append(frame)

#append frame1 to list of all dataframes already made
dataframe_names.append(frame1)

#concatenate all dataframes into on
df = pd.concat(dataframe_names, axis=0, ignore_index=True)
display(df)
```
